```python
import pandas as pd
import folium

# CSV
data = pd.read_csv('port_.csv')

# 
m = folium.Map(location=[20,0], zoom_start=2)  # [20, 0]

# 
for index, row in data.iterrows():
    folium.Marker(
        location=[row['latitudeDecimal'], row['longitudeDecimal']],
        icon=folium.Icon(icon='cloud'),
        popup=f"Port: {row['port']}, Country: {row['country_name']}"
    ).add_to(m)
    
    # 30km
    folium.Circle(
        location=[row['latitudeDecimal'], row['longitudeDecimal']],
        radius=17500,  # 30km
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.3
    ).add_to(m)

# HTML
m.save('ports_map.html')
m
```
