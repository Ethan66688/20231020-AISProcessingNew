```python
import pandas as pd
import folium

# 从CSV文件中读取数据
data = pd.read_csv('port_测试.csv')

# 创建一个基于世界地图的地图对象
m = folium.Map(location=[20,0], zoom_start=2)  # 使用经纬度[20, 0]作为初始中心，可以根据你的数据集进行调整

# 将港口的经纬度添加到地图上
for index, row in data.iterrows():
    folium.Marker(
        location=[row['latitudeDecimal'], row['longitudeDecimal']],
        icon=folium.Icon(icon='cloud'),
        popup=f"Port: {row['port']}, Country: {row['country_name']}"
    ).add_to(m)
    
    # 为每个港口位置画一个30km的圆形
    folium.Circle(
        location=[row['latitudeDecimal'], row['longitudeDecimal']],
        radius=17500,  # 30km
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.3
    ).add_to(m)

# 保存为HTML文件
m.save('ports_map.html')
m
```
